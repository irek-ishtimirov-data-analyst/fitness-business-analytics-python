### Задача 3. 
Преобразуйте код выше в функцию, которая берет на вход список и пороговое значение для слишком больших значений и расчитывает среднее значение списка, исключая слишком большие значения

```python
def calc_avg(lst, treshold):
    sum_pay = 0
    cnt_pay = 0
    for i in lst:
        try:
            if i < treshold:
                sum_pay += i
                cnt_pay += 1
        except TypeError:
            print(f'{i} - нечисловое значение')
    try:
        avg_pay = sum_pay / cnt_pay
        return round(avg_pay, 2)
    except ZeroDivisionError:
        print(f'Список не должен быть пустым или должен содержать хотя бы одно значение меньше {treshold}')
```

```python
calc_avg(pay_lst, 15000)
```

```python
calc_avg(lst1, 4000)
```

# Задача 5.
Преобразуйте код из задачи 4 в функцию, которая принимает на вход только список и возвращает среднее значение рассчитанное после исключения выбросов

```python
def calc_avg_advanced(lst):
    try:
        q1 = np.percentile(lst, 25)
        q3 = np.percentile(lst, 75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        sum_pay = 0
        cnt_pay = 0
        for i in lst:
            if (i > lower_bound) and (i < upper_bound):
                sum_pay += i
                cnt_pay += 1
        avg_pay = sum_pay / cnt_pay                  #
        return round(avg_pay, 2)
    except:
        print(f'Список не должен содержать нечисловые значения или быть пустым!')
        
```

```python
calc_avg_advanced(pay_lst)
```

```python
lst2 = [1000, 2000,3000, 'a']
calc_avg_advanced(lst2)
```

### Задача 6.

```python
pay_month =  {'jan':[6400, 8000, 18000, 3600, 36000, 1600], 'feb': [1600, 800, 800, 3200, 5800, 5600],\
              'march': [1600, 800, 8000, 18000, 9000, 3200]}
```

```python
avg_pay = {}
for i in pay_month.keys():
    pay = pay_month[i]
    avg = sum(pay) / len(pay)
    avg_pay[i] = round(avg, 2)
```

```python
avg_pay
```

```python
def calc_avg_dict(mydict):
    avg_pay = {}
    for i in mydict.keys():
        try:
            pay = mydict[i]
            avg = sum(pay) / len(pay)
            avg_pay[i] = round(avg, 2)
        except:
            avg_pay[i] = None
    return avg_pay
```

```python
calc_avg_dict(pay_month)
```

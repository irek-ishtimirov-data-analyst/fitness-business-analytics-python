**Задача 1.** Рассчитайте средний возраст по всему списку age_lst

**Задача 5 (со звёздочкой).** Оберните предыдущий код в функцию, которая принимает на вход список и рассчитывает средний возраст, исключая из расчётов выбросы

```python
import numpy as np
```

```python
age_lst = [23, 45, 18, 80, 22, 25, 34, 37, 41, 88, 19, 20, 18, 45, 41, 39, 31, 32, 22, 28, 37]
```

##### Задача 1. Рассчитайте средний возраст по всему списку age_lst

```python
# Расчёт среднего возраста посетителей.

avg_age = sum(age_lst) / len(age_lst)
print(f'Средний возраст посетителей {round(avg_age)} лет.')     
```

##### Задача 2. Рассчитайте средний возраст для посетителей, которые моложе 50 лет

```python
# Расчёт среднего возраста посетителей моложе 50 лет.

sum_age = 0  # Создаём две переменные sum_age и cnt_age
cnt_age = 0
for i in age_lst: # Запускаем цикл, где i принимает каждое значение из списка age_lst
    if i < 50:    # Если значение, принявшее i меньше 50
        sum_age += i # Тогда это значение прибавляется к предыдущему значению в переменной sum_age
        cnt_age += 1 # В переменную cnt_age прибавляется единица
avg_age = sum_age / cnt_age # Сумму возрастов делим на сумму единиц и получаем средний возраст
print(f'Средний возраст посетителей моложе 50 лет, составляет {round(avg_age)} лет.')        
```

##### Задача 3. Оберните предыдущий код в функцию, которая на вход принимает список и пороговое значение

```python
# Объявление функции, которая рассчитывает средний возраст посетителей моложе 50 лет.

def calc_avg_age(lst, treshold):
    sum_age = 0
    cnt_age = 0
    for i in age_lst:
        if i < treshold:
            sum_age += i
            cnt_age += 1
    avg_age = sum_age / cnt_age
    return round(avg_age)
```

```python
calc_avg_age(age_lst, 50)
```

##### Задача 4. Рассчитайте среднее значение для всех элементов списка, за исключением тех, которые лежат вне интервала [Q1-1,5IQR; Q3+1,5IQR]

```python
print(np.mean(age_lst).round())
```

```python
print(np.median(age_lst))
```

```python
# Разбиение списка на квартили.

q1 = np.percentile(age_lst, 25)
q2 = np.percentile(age_lst, 50)
q3 = np.percentile(age_lst, 75)
print(q1)
print(q2)
print(q3)
```

```python
# Расчёт межквартильного размаха, нижней и верхней границы

iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
print(iqr)
print(lower_bound)
print(upper_bound)
```

```python
# Расчёт среднего возраста посетителей с использованием нижней и верхней границы.

sum_age = 0
cnt_age = 0
for i in age_lst:
    if (i > lower_bound) and (i < upper_bound):
        sum_age += i
        cnt_age += 1
avg_age = sum_age / cnt_age
print(f'Средний возраст {round(avg_age)} лет.')
```

##### Задача 5 (со звёздочкой). Оберните предыдущий код в функцию, которая принимает на вход список и рассчитывает средний возраст, 
##### исключая из расчётов выбросы

```python
# Объявление функции, которая рассчитывает средний возраст посетителей не включая в расчёты выбросы.

def calc_avg_age_advanced(lst):
    q1 = np.percentile(age_lst, 25)
    q3 = np.percentile(age_lst, 75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    sum_age = 0
    cnt_age = 0
    for i in age_lst:
        if (i > lower_bound) and (i < upper_bound):
            sum_age += i
            cnt_age += 1
    avg_age = sum_age / cnt_age
    return round(avg_age)
```

```python
average_age = calc_avg_age_advanced(age_lst)
print(f'Средний возраст посетителей {average_age} лет.')
```
